# Tiny Neural Network using The Autograd Engine 

## Importing our autograd engine as a library to use it

```python
from autograd import *
```

## The model

We use one neuron:
y=w⋅x+b

Where:
- x → input (data, fixed)
- w → weight (parameter)
- b → bias (parameter)
- y → prediction

```python
w = Node(0)
b = Node(0)
```

```python
x = 2
y_true = 7
```

```python
y_pred = w*x + b
```

```python
loss = (y_pred-y_true)**2
```
